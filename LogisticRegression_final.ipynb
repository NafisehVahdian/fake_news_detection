```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from scipy.sparse import coo_matrix
# Load the data
X_train = np.load("/content/X_train.npy", allow_pickle=True).item()
X_test = np.load("/content/X_test.npy", allow_pickle=True).item()
X_val = np.load("/content/X_val.npy", allow_pickle=True).item()
y_train = np.load("/content/y_train.npy", allow_pickle=True)
y_test = np.load("/content/y_test.npy", allow_pickle=True)
y_val = np.load("/content/y_val.npy", allow_pickle=True)


# Ensure all sparse matrices are converted to dense
if isinstance(X_train, coo_matrix):
    X_train = X_train.toarray()
if isinstance(X_test, coo_matrix):
    X_test = X_test.toarray()
if isinstance(X_val, coo_matrix):
    X_val = X_val.toarray()


# Step 2: Train the logistic regression model
model = LogisticRegression(max_iter=1000, random_state=42)
model.fit(X_train, y_train)

# Step 5: Validate the model
if isinstance(X_val, coo_matrix):
    X_val = X_val.toarray()
y_val_pred = model.predict(X_val)
print("Validation Accuracy:", accuracy_score(y_val, y_val_pred))
print("\nClassification Report (Validation):")
print(classification_report(y_val, y_val_pred))

# Step 4: Test the model
y_test_pred = model.predict(X_test)
print("Test Accuracy:", accuracy_score(y_test, y_test_pred))
print("\nClassification Report (Test):")
print(classification_report(y_test, y_test_pred))
```
